# 风电功率数据探索与可视化

本notebook用于SCADA风机数据的初步探索与可视化。

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/wtbdata_245days.csv')
df.head()
```

```python
# 缺失值统计
df.isnull().sum()
```

```python
# 单台风机功率曲线
turbid = 1
df1 = df[df['TurbID'] == turbid]
plt.figure(figsize=(16,4))
plt.plot(df1['Tmstamp'], df1['Patv'])
plt.title(f'Turbine {turbid} Power Output')
plt.xlabel('Time')
plt.ylabel('Power (kW)')
plt.show()
```
